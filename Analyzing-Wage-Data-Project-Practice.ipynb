```python
import pandas as pd
df_wages = pd.read_excel('Wages_project.xlsx')

df_wages.head()
```

```python
# renaming columns
column_mapper = {
    'Occupation title (click on the occupation title to view its profile)' : 'Occupation title'
}

df_wages = df_wages.rename(
    mapper=column_mapper,
    axis=1)

df_wages.head()
```

```python
#dropping unnecessary columns

drop_columns = ['Occupation code', 'Level', 'Employment RSE', 'Mean wage RSE']

df_wages = df_wages.drop(
    labels=drop_columns,
    axis=1)

df_wages.head()
```

```python
df_wages.info()
```

```python
# mean and median wage comparrison
median_hr_wgs = df_wages['Median hourly wage']

Mean_hr_wgs = df_wages['Mean hourly wage']

mm_dif = median_hr_wgs - Mean_hr_wgs

df_wages['mm_dif'] = mm_dif

df_wages[['Occupation title', 'Median hourly wage', 'Mean hourly wage', 'mm_dif' ]].head()
```

```python
amw = df_wages['Annual mean wage']

diff_perc = (mm_dif / amw) * 100

df_wages['diff_perc'] = diff_perc

wage_diff = df_wages[['Occupation title','Employment', 'Median hourly wage', 'Mean hourly wage', 'mm_dif', 'diff_perc']]

wage_diff.head()
```

```python
asc_wage_diff = wage_diff.sort_values(by='Employment', ascending= True)

asc_wage_diff.head()
```

```python
desc_asc_wage_diff = wage_diff.sort_values(by='Employment', ascending= False)

desc_asc_wage_diff.head()
```
